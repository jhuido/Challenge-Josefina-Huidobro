# AICI Challenge

Your task involves categorizing coordinates from CAD files into cardinal directions based on given examples.

Given the three cases below, you are required to classify these lists into a dictionary categorized into 4 keys: W, N, E, S. Here are the expected outputs for each example:

**Example 1:**
- **Input:** `array: [(-2, -2), (-2, 2), (2, 2), (2, -2), (-2, -2)]`
- **Output:** `{"W": [(-2, -2), (-2, 2)], "N": [(-2, 2), (2, 2)], "E": [(2, 2), (2, -2)], "S": [(2, -2), (-2, -2)]}`

**Example 2:**
- **Input:** `array: [(0, 0), (-1, 4), (7, 6), (7.25, 5), (4.25, 4.25), (5, 1.25), (0, 0)]`
- **Output:** `{"W": [(0, 0), (-1, 4)], "N": [(-1, 4), (7, 6)], "E": [(7, 6), (7.25, 5), (4.25, 4.25), (5, 1.25)], "S": [(5, 1.25), (0, 0)]}`

**Example 3:**
- **Input:** `array: [(0, 0), (0, 10), (20, 10), (20, -3), (15, -3), (15, 2), (10, 2), (10, 0), (0, 0)]`
- **Output:** `{"W": [(0, 0), (0, 10)], "N": [(0, 10), (20, 10)], "E": [(20, 10), (20, -3)], "S": [(20, -3), (15, -3), (15, 2), (10, 2), (10, 0), (0, 0)]}`

```python
"""
We added this code for you to visualize the plots of the points in the cases.
"""
import matplotlib.pyplot as plt

case1 = [(-2, -2), (-2, 2), (2, 2), (2, -2), (-2, -2)]
case2 = [(0, 0), (-1, 4), (7, 6), (7.25, 5), (4.25, 4.25), (5, 1.25), (0, 0)]
case3 = [(0, 0), (0, 10), (20, 10), (20, -3), (15, -3), (15, 2), (10, 2), (10, 0), (0, 0)]

fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()
fig3, ax3 = plt.subplots()

ax1.plot([point[0] for point in case1], [point[1] for point in case1], marker='o')
ax2.plot([point[0] for point in case2], [point[1] for point in case2], marker='o')
ax3.plot([point[0] for point in case3], [point[1] for point in case3], marker='o')

ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Points of Case 1')

ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Points of Case 2')

ax3.set_xlabel('X')
ax3.set_ylabel('Y')
ax3.set_title('Points of Case 3')

plt.show()
```

```python
       
def calculateMedia(index,array):
    values=[]
    for i in array:
        values.append(i[index])
    media=min(values)+((max(values)-min(values))/2)
    return media

def addCoord(element,array):
    if element not in array:
        array.append(element)

def sideDetection(case):
       
    if(len(case)==0):
        return 
    
    x_media=calculateMedia(0,case)
    y_media=calculateMedia(1,case)
    
    output={}
    w=[]
    e=[]
    s=[]
    n=[]
    
    for coord in case:
        if(coord[0]<x_media):
            addCoord(coord,w)
        else:
            addCoord(coord,e)
        if(coord[1]<y_media):
            addCoord(coord,s)
        else:
            addCoord(coord,n)
            
    output["W"]=w
    output["N"]=n
    output["E"]=e
    output["S"]=s
    
    return output

case1 = [(-2, -2), (-2, 2), (2, 2), (2, -2), (-2, -2)]
case2 = [(0, 0), (-1, 4), (7, 6), (7.25, 5), (4.25, 4.25), (5, 1.25), (0, 0)]
case3 = [(0, 0), (0, 10), (20, 10), (20, -3), (15, -3), (15, 2), (10, 2), (10, 0), (0, 0)]

print(sideDetection(case1))
print(sideDetection(case2))
print(sideDetection(case3))
```
